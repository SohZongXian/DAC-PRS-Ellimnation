```python
sample = pd.read_csv("Sample Submission DAC.csv")
train = pd.read_csv("Train.csv")
test = pd.read_csv("Test.csv")

## Set column name as index and drop
train = train.set_index('Unnamed: 0')
train = train.reset_index(drop=True)
test = test.set_index('Unnamed: 0')
test = test.reset_index(drop=True)
```

```python
##Convert both date columns into datetime format
train['dealing_date'] = pd.to_datetime(train['dealing_date'])
train['buying_date'] = pd.to_datetime(train['buying_date'])

test['dealing_date'] = pd.to_datetime(test['dealing_date'])
test['buying_date'] = pd.to_datetime(test['buying_date'])
```

```python
train.dtypes
```

```python
## Identify missing values
train.isnull().sum()
```

```python
## fill in null value
train['distance']=train['distance'].fillna(0)
test['distance']=test['distance'].fillna(0)
```

```python
## Create new column "Number of days" in train and test df
train['Number of days'] = (train['dealing_date'] - train['buying_date']).dt.days
test['Number of days'] = (test['dealing_date'] - test['buying_date']).dt.days

##Drop cnt, dealing and time_date columns
del train['time_date']
del train['dealing']

del test['time_date']
```

```python
## Fill missing values in number of days with 0
train['Number of days'] = train['Number of days'].fillna(0)
```

```python
## Convert float to int
train['Number of days'] = train['Number of days'].astype(int)
test['Number of days'] = test['Number of days'].astype(int)
```
